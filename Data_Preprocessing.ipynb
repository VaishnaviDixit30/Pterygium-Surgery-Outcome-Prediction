```python
import pandas as pd
df = pd.read_csv("original_data.csv")
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib
df["gender"] = df["gender"].map({"Male": 0, "Female": 1})
df["recurrent"] = df["recurrent"].map({"No": 0, "Yes": 1})
label_cols = [ 'area', 'affected_eye','pterygium_location','followup1_complication', 'followup2_complication']

label_encoders = {}
for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
    joblib.dump(le, f"label_encoder_{col}.pkl")

gender_map = {"Male": 0, "Female": 1}
recurrent_map = {"No": 0, "Yes": 1}
joblib.dump(gender_map, "map_gender.pkl")
joblib.dump(recurrent_map, "map_recurrent.pkl")
```

```python
non_feature_cols = ["patient_id","surgery_date","followup1_date","followup2_date","surgery_type"]
categorical_cols = label_cols + ["gender", "recurrent"]
numerical_cols = [col for col in df.columns if col not in categorical_cols + non_feature_cols]
numerical_cols
```

```python
df_unscaled = df.copy()
df_unscaled.to_csv("spd_unscaled.csv", index=False)
```
